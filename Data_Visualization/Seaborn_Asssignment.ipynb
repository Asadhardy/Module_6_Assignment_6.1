```python
'''
Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.
Answer:
    generating a synthetic dataset with correlated features and visualize the correlation matrix using a Bokeh heatmap
    the codes are as following
'''

# generating a synthetic dataset
np.random.seed(42)
n_samples = 100
X = np.random.rand(n_samples, 1)
Y = 2 * X + np.random.normal(0, 0.1, size=(n_samples, 1))  # Correlated with X
Z = -X + np.random.normal(0, 0.1, size=(n_samples, 1))  # Inversely correlated with X

# creating a DataFrame
data = pd.DataFrame(np.hstack((X, Y, Z)), columns=['X', 'Y', 'Z'])

# calculating the correlation matrix
correlation_matrix = data.corr()

# creating a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})
plt.title("Correlation Matrix Heatmap")
plt.show()


'''
Insights:
    This heatmap shows how the features in the synthetic dataset are related to each other. Darker colors mean stronger relationships, 
    and lighter colors mean weaker ones. A positive correlation (close to +1) means that when one feature increases, the other usually does too. 
    A negative correlation (close to -1) means that when one feature goes up, the other tends to go down. You can see a strong relationship between
    X and Y, which makes sense since they are directly linked. Z has moderate relationships with both X and Y, showing some indirect connections.
'''
plt.show()
```
