```python
'''
Q1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three
    dimensional space.
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

Answer:
    To generate a 3d scatter plot to visualize the distribution of data points in a three dimensional space using the given dataset.
    The code use the matplotlib's 3d plotting capabilities.
Answer:
    To generate a 3D scatter plot using the given dataset, into Plotly.
    3d Scatter Plot Using Plotly
    The code is as following.
'''

# Set random seed for reproducibility
np.random.seed(30)

# Generate the dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

# Create a 3d scatter plot
fig = px.scatter_3d(df, x="X", y="Y", z="Z", title="3D Scatter Plot of Randomly Generated Data", opacity=0.7)


'''
Explanation:
    3D Scatter Plot
    We use plotly.express.scatter_3d() to generate a 3D scatter plot.
    x, y, and z parameters specify the three axes of the plot.
    opacity=0.7 adjusts the transparency to make overlapping points easier to visualize.
Insights:
    This approach will first will show the distribution of data points in the form of 3D scatter plot
    in a three dimensional space.
    With the help of plotly we can interact with the dataPoints 
'''
# Showing the plot
fig.show()
```

```python
'''
Q3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
Answer:
    To generate a heatmap that visualizes the variation in sales across different months and days using the given sales data.
    performing using Plotly is as following:
'''
# Generate the dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(["Jan", "Feb", "Mar", "Apr", "May"], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivoting the data to create a matrix-like format suitable for heatmaps
heatmap_data = df.pivot_table(index="Day", columns="Month", values="Sales", aggfunc="mean")

# Creating the heatmap using Plotly
fig = px.imshow(heatmap_data, labels=dict(x="Month", y="Day", color="Sales"), 
                title="Sales Variation Across Months and Days")


'''
Insight:

    The heatmap is a visual tool that helps to quickly spot patterns or variations in sales across different months and days. 
    It uses color intensity to represent the magnitude of sales, making it easy to identify periods of higher or lower sales at a glance. 
    By grouping the data by days and months, the heatmap allows us to observe any temporal trends or anomalies in sales distribution over time. 
    Plotly adds interactivity, making the exploration of data points more dynamic and intuitive.
'''
fig.show()
```

```python
'''
Q4. Using the given x and y data, generate a 3D surface plot to visualize the function
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()}
df = pd.DataFrame(data)
Answer
To generate a 3D surface plot for the function z = sin(sqrt(x^2 + y^2)) using the given x, y, and z data.
'''

# Generate the x, y, and z data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays and create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3d surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Add title and labels
fig.update_layout(title="3D Surface Plot of z = sin(sqrt(x^2 + y^2))",
                  scene=dict(
                      xaxis_title="X",
                      yaxis_title="Y",
                      zaxis_title="Z"
                  ))


'''
Insight:
    done with the help of gpt, only for understanding and exlanations.
    
    The 3D surface plot of z = sin(sqrt(x + y)) illustrates a wave-like pattern created by concentric ripples based on the distance from
    the origin, effectively showcasing the function's behavior in three-dimensional space with peaks and troughs representing maximum 
    and minimum values.
    '''
# Showing the plot
fig.show()
```

```python
'''
Q5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.
np.random.seed(25)
data = {
    'Country': ["USA", "Canada", 'UK',
'Germany', 'France'],
    'Population':
np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000,5)}
df = pd.DataFrame(data)
Answer:
    Creating a bubble chart using Plotly, where each country's population is on the y-axis, GDP is on the x axis, 
    and the bubble size is proportional to the population.
'''

# Generates the dataset
np.random.seed(25)
data = {
    'Country': ["USA", "Canada", "UK", "Germany", "France"],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Createing a bubble chart using Plotly
fig = px.scatter(df, 
                 x="GDP", 
                 y="Population", 
                 size="Population", 
                 hover_name="Country", 
                 title="Bubble Chart: Country Population vs GDP",
                 labels={"GDP": "GDP (in billion $)", "Population": "Population (in millions)"})


'''
Insight:

    This bubble chart shows how each country's GDP relates to its population, with bubble sizes representing population numbers. 
    You can quickly spot that bigger bubbles mean larger populations, which helps compare countries easily. For example, if one 
    country has a high GDP but a small bubble, it suggests its economic power is greater than what you'd expect from its population size, 
    which is an interesting observation.
'''
# Show the plot
fig.show()
```
