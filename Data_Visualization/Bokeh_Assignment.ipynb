```python

'''
Module_6
Bokeh_Assignment
'''
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis256
import warnings
warnings.filterwarnings('ignore')
```

```python
'''
Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.?
Answer:
    You can create a sine wave plot using Bokeh with x values rnging from 0 to 10 and
    y values as the sine ofx
'''
# generating the output in a notebook
output_notebook()

# generating data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a new plot
plot = figure(title = "Sine Wave", y_axis_label = "sin(x)", x_axis_label = "x",
             width = 800, height = 400)

# Adding a line rendered with legent and line thickness
plot.line(x, y , legend_label= "sim(x)", line_width = 2)

# showing the plot
show(plot)
'''
Insights:
    This Bokeh plot shows a sine wave with x values from 0 to 10, and the y values come from the sine function 
    sin(x). 
    What's cool about using Bokeh is that you can interact with the plot—hover over it to see exact numbers or zoom in and out to get a better
    view.k.


'''
```

```python

'''
2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.?
Answer:
    To create a Bokeh scatter plot using randomly generated x and y values,
    and customize the marker sizes and colors based on additional columns, the code is as following:
'''


# To display the output in the notebook
output_notebook()

# Generating random data
np.random.seed(42)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
sizes = np.random.uniform(10, 50, 100)
colors = np.random.uniform(0, 100, 100)

# Creating a DataFrame
data = pd.DataFrame({"x": x, "y": y, "sizes": sizes, "colors": colors})

# Convert the DataFrame into a Bokeh ColumnDataSource
source = ColumnDataSource(data)

# Create a new scatter plot
plot = figure(title="Bokeh Scatter Plot with Varying Sizes and Colors", x_axis_label='X-axis', y_axis_label='Y-axis', width=800, height=400)

# Map the colors column to the colors of the scatter points
color_mapper = linear_cmap(field_name="colors", palette=Viridis256, low=min(colors), high=max(colors))

# adding circles glyphs (scatter markers)
plot.circle(x="x", y="y", size="sizes", color=color_mapper, fill_alpha=0.6, line_color="black", source=source)


'''
Insight:
    In this scatter plot, x and y values are randomly generated using numpy. The marker sizes are customized based 
on the sizes column, and colors are mapped based on the colors column using Bokeh's linear_cmap function. The Viridis256 palette is used
for coloring the markers, and each point varies in size and color to make the plot visually distinc
'''
# Show the plot
show(plot)
```

```python
'''
Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ["Apples", "Oranges", "Bananas", "Pears"]
counts = [20, 25, 30, 35]
Answer:
     generate a Bokeh bar chart for the given dataset of fruits and their counts the 
     code is as following:
'''
# Displaying the output in a notebook
output_notebook()

# dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Creating a new bar chart
plot = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', width=800, height=400)

# Adding bars to the chart
plot.vbar(x=fruits, top=counts, width=0.9)

'''
insights:
    This Bokeh bar chart shows the counts of different fruits, making it easy to compare how many of each there are. 
    Each bar represents a fruit, and its height shows the count, so you can quickly see which fruit has the most or least. Plus, 
    with Bokeh's interactive features, you can hover over the bars to get more information.
'''
# displaying the plot
show(plot)
```

```python
'''
Q4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
Answer:
     Create a Bokeh histogram to visualize the distribution of the given data the codes are as following:
'''
# Displaying the output in a notebook
output_notebook()

# Generating random data
data_hist = np.random.randn(1000)

# Creating the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Creating a new plot for the histogram
plot = figure(title="Histogram of Random Data", 
              x_axis_label="Value", 
              y_axis_label="Frequency", 
              width=800, 
              height=400)

# Adding bars to the histogram
plot.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="blue", line_color="black")

# Displaying the plot

'''
insights:
    This histogram visualizes how the random data is distributed across different value ranges. 
    Each bar represents the frequency of data points that fall within specific intervals. It’s a great way to see patterns 
    in the data, like where most values cluster and the overall shape of the distribution. With Bokeh, you can also interact 
    with the plot, making it easier to analyze the distribution visually!
'''
show(plot)
```

```python
'''
Q5. Create a Bokeh heatmap using the provided dataset?
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
Answer:
    Creating a Bokeh heatmap using the provided dataset. the code are as following
'''
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis256

# Displaying the output in a notebook
output_notebook()

# Generating random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Creating a ColumnDataSource from the heatmap data
source = ColumnDataSource(data={"x": xx.flatten(), 'y': yy.flatten(), "value": data_heatmap.flatten()})

# Defining the low and high values for color mapping
low = data_heatmap.min()
high = data_heatmap.max()

# Creating a new figure for the heatmap
plot = figure(title="Heatmap", 
              x_axis_label="X-axis", 
              y_axis_label="Y-axis", 
              width=800, 
              height=400)

# Using the rect function to create the heatmap
plot.rect(x="x", y='y', width=0.1, height=0.1, source=source, 
          fill_color=linear_cmap("value", Viridis256, low=low, high=high), line_color="black")

'''
Insights:
    This heatmap shows the values from the dataset on a grid, where the color intensity gives you an idea of how big the values are. 
    The rect function is used to create colored squares for the heatmap, making it simple to spot patterns and changes in the data. 
    Plus, Bokeh's interactive features let you hover over the squares to see exact values, adding an extra layer of insight to your visualization!
'''
# Displaying the plot
show(plot)
```
